## grouping

```python
sns.relplot(x='tip', y='total_bill', data=tips, hue='smoker') # huee is the colour sepration terms in seaborn in relational plot
```

```python
sns.relplot(x='tip', y='total_bill', data=tips, hue='day')
```

```python
sns.relplot(x='tip', y='total_bill', data=tips, hue='day', style='smoker') # this is group in group, # style bill give the grouping by its own idivisual style of represantaion
```

```python
sns.relplot(x='tip', y='total_bill', data=tips, hue='day', style='smoker', size='size') # size is also gives on grouping
```

```python
sns.relplot(x='tip', y='total_bill', data=tips, size='tip')# differentiate on the basis of the tips
```
