```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame(np.random.randn(10,4), index=pd.date_range('2021-07-18', periods=10), columns=['a','b','c','d'])
```

```python
df
```

```python
df['new']=df['c']+df['d']
```

```python
df
```

```python
df['new_col_square_d']=df['d'].apply(lambda x: x**2)
```

```python
df
```

```python
df['new_col_square_d_log']=df['d'].apply(lambda x: np.log(x)) # apply the lambda function
```

```python
df
```

```python
def text(x):
    if (x<0):
        return x
    else:
        return np.log(x)
    
```

```python
df['new_col_square_d_log_1']=df['d'].apply(text)
```

```python
df
```

```python
def test(x):
    if x<0:
        return 1
    else:
        return 0
```

```python
df['new_a']=df['a'].apply(test) # apply the function in the columns
```

```python
df
```

```python
df.sort_values(by='b') # its sorting of the columns
```

```python
df['New_date']=df.index # create the day
```

```python
df
```

```python
df['day']=df.index.day # you can also do with the split method by ('-')
df['month']=df.index.month
df['year']=df.index.year
```

```python
df
```

```python
df['day_name']=df.index.day_name() # know the day name
```
