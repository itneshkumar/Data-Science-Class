```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
sns.relplot(x='tip', y='total_bill', data=tips)
```

```python
sns.relplot(x='tip', y='total_bill', data=tips, hue='smoker') # huee is the colour sepration terms in seaborn in relational plot
```

```python
sns.relplot(x='tip', y='total_bill', data=tips, hue='day')
```

```python
sns.relplot(x='tip', y='total_bill', data=tips, hue='day', style='smoker') # style bill give the grouping by its own idivisual style of represantaion
```
